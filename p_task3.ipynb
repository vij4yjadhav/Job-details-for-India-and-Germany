```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, time
```

```python

data = {
    'Country': ['India', 'Germany', 'India', 'Germany', 'India', 'Germany'],
    'Job Title': ['Data Scientist', 'Data Scientist', 'Art Teacher', 'Art Teacher', 'Aerospace Engineer', 'Aerospace Engineer'],
    'Salary': [25000, 20000, 12000, 18000, 45000, 40000],
    'Experience': [3, 4, 5, 3, 6, 7],
    'Job Posting Date': ['07/01/2023', '07/15/2023', '06/20/2023', '07/25/2023', '07/10/2023', '07/05/2023']
}

df = pd.DataFrame(data)
```

```python
df = df[(df['Country'].isin(['India', 'Germany'])) &
        (df['Experience'] > 2) &
        (df['Salary'] > 10000) &
        (df['Job Posting Date'] < '08/01/2023')]
```

```python
current_time = datetime.now().time()
start_time = time(15, 0)  
end_time = time(18, 0)    
```

```python
if start_time <= current_time <= end_time:

    plt.figure(figsize=(10, 6))
    
     
    colors = {'India': 'orange', 'Germany': 'green'}
    
    for country, color in colors.items():
        country_data = df[df['Country'] == country]
        plt.bar(country_data['Job Title'], country_data['Salary'], color=color, label=country)
    
    plt.xlabel('Job Title')
    plt.ylabel('Salary ($)')
    plt.title('Job Details for India and Germany (B.Tech, Full Time, Experience > 2 years)')
    plt.legend()
    plt.show()
else:
    print("The chart is only available between 2 PM IST to 5 PM IST.")
```
